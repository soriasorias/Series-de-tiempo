```python
# Leer datos de los archivos
import numpy as np
import random
import matplotlib.pyplot as plt
import math

serie1 = []
serie2 = []
serie3 = []

f1 = open("SerieDeTiempoUniforme.txt", "r")
a1 = f1.read()
valores1 = a1.split(',')
for i in valores1:
    if i == "":
        serie1.append(1)
    else:
        serie1.append(int(i))
f1.close()

f2 = open("SerieDeTiempoGauss.txt", "r")
a2 = f2.read()
valores2 = a2.split(',')
for i in valores2:
    if i == "":
        serie2.append(1)
    else:
        serie2.append(float(i))
f2.close()

f3 = open("SerieDeTiempoGamma.txt", "r")
a3 = f3.read()
valores3 = a3.split(',')
for i in valores3:
    if i == "":
        serie3.append(1)
    else:
        serie3.append(float(i))
f3.close()

datos_uniformes = np.array(serie1)
datos_gaussiana = np.array(serie2)
datos_gamma = np.array(serie3)
```

```python
datos_gaussiana_tf = []
datos_gamma_tf = []

for i in serie2:
    f = i*125+500
    datos_gaussiana_tf.append(f)
for i in serie3:
    f = i*80
    datos_gamma_tf.append(f)

intervalo_sturges = int(math.log2(len(datos_uniformes_tf)))
plt.hist(datos_gamma_tf,intervalo_sturges,color='#22AB9D', rwidth=0.90);
```
