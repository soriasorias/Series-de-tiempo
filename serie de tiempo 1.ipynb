```python
import random
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
n = 10000
serie = []
for i in range (1,n):
    q = random.randint(2, 1000)
    serie.append(q)
```

```python
f = open("SerieDeTiempoUniforme.txt", "w")

for i in serie:
    f.write(str(i)+",")

f.close()
```

```python
intervalo = 1000
fig, ax = plt.subplots()
plt.hist(serie,intervalo,color='#22AB9D', rwidth=0.90)
plt.title('Histograma de números aleatorios.')
plt.xlabel('Número')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
intervalo = int(math.log2(n))
print(intervalo)
fig, ax = plt.subplots()
plt.hist(serie,intervalo,color='#22AB9D', rwidth=0.90)
plt.title('Histograma de números aleatorios. Sturges')
plt.xlabel('Número')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
s = np.std(serie)
intervalo = int((3.5 * s * n ** (-1/3)))
print(intervalo)
fig, ax = plt.subplots()
plt.hist(serie,intervalo,color='#22AB9D', rwidth=0.90)
plt.title('Histograma de números aleatorios. Scott')
plt.xlabel('Número')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
s = np.std(serie)
serieord = sorted(serie)
lista1 = serieord[1:5000]
lista2 = serieord[5001:n]
upp = sum(lista2)/len(lista2)
low = sum(lista1)/len(lista1)
intervalo = int(2*(upp - low)/(n ** (1/3)))
print(intervalo)
fig, ax = plt.subplots()
plt.hist(serie,intervalo,color='#22AB9D', rwidth=0.90)
plt.title('Histograma de números aleatorios. Freedman–Diaconis')
plt.xlabel('Número')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```
